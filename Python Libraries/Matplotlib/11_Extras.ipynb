```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
```

```python
x = np.arange(0, 10)
y = np.arange(10, 20)

plt.plot(x, y)
plt.savefig('Test1.png')
```

```python
x1 = np.arange(0, 5 * np.pi, 0.1)
y1 = np.sin(x1)
y2 = np.cos(x1)

plt.subplot(2,1,1)
plt.plot(x1, y1, 'g--')
plt.title('Sine wave')

plt.subplot(2,1,2)
plt.plot(x1, y2, 'r*--')
plt.title('Cosine wave')

plt.tight_layout()
```

```python
#Histograms -- a histogram is a graph that shows the distribution of numerical data.
#              It takes continuous numbers, groups them into consecutive, non-overlapping intervals (called bins),
#              and uses bars to show how many data points fall into each bin.
```

```python
#Box-Plots

data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, vert=True, patch_artist=True)
plt.tight_layout()
```

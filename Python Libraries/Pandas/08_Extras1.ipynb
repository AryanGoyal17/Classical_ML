```python
import numpy as np 
import pandas as pd 
```

```python
#1--Forming a dataframe..

df = pd.DataFrame(np.arange(0, 30).reshape(5,6), index = ["Row1", "Row2", "Row3", "Row4", "Row5"], columns = ["Column1", "Column2", "Column3", "Column4", "Column5", "Column6"])
print(df)
```

```python
#2-- 
df.to_csv("Test1.csv")
```

```python
#3-- 

print(df.loc["Row1"])
type(df.loc["Row1"])

print(df.iloc[0:3, 3:])
type(df.iloc[0:3, 3:])
```

```python
#4-- Converting dataframe to np array

print(df.iloc[:,:].values)
type(df.iloc[:,:].values)
```

```python
#5-- Checking if there are any missing values in the dataframe--

print(df.isnull().sum())

#If there were missing values, they can be handled by df.fillna or df.dropna
```

```python
#6-- To find the unique categories in a column

#M1--

print(df["Column3"].value_counts())

#M2--

print(df["Column3"].unique())
type(df["Column3"].unique())
```

```python
#8(a)

# df_3 = pd.read_csv("Test3.csv") #Shows error

df_3 = pd.read_csv("Test3.csv", sep = ";")
print(df_3)
```

```python
#8(e)

df_4 = pd.read_csv("Test3.csv", sep = ";", index_col="Unnamed: 0")
print(df_4) #USING index_col
```

```python
#8(b)

from io import StringIO , BytesIO

data = ("Column1,Column2,Column3\n"
         "1,2,3\n"
         "a,b,c\n")
print(type(data))
df_5 = pd.read_csv(StringIO(data))
print(df_5)
```

```python
#8(c)

df_6 = pd.read_csv(StringIO(data), usecols=lambda x : x.upper() in ['COLUMN1', 'COLUMN3']) #same as usecols = ['Column1', 'Column3']
print(df_6) #for reading only certain columns..
```

```python

df_7 = pd.DataFrame(np.arange(0,20).reshape(5,4))
print(df_7)
df_7.to_csv("Test2.csv")
```

```python
#8(d)-- Specifying columns data types..

df_8 = pd.read_csv("Test2.csv", dtype = float)
df_8 = df_8.drop(columns=['Unnamed: 0'])
print(df_8)
```

```python
df_9 = pd.read_csv("Test2.csv", dtype = {0: int, 1: float, 2:'Int64'})
df_9 = df_9.drop(columns=["Unnamed: 0"]) #0 -- got dropped that;s why first column has dtype = float..
print(df_9, end = "\n\n")
print(df_9.dtypes)
```

```python
#8(e) -- Use of index_col = False

data1 = ('a,b,c\n'
        '1,2,3,\n'
         '4,5,6,')

df_10 = pd.read_csv(StringIO(data1))
print(df_10, end = "\n\n")

df_11 = pd.read_csv(StringIO(data1), index_col=False)
print(df_11)
```

```python
#8(f) -- Quoting and Escape characters..

data2 = ('a,b\n"Hello \\"Bob\\" nice to see you", 5\n')

df_12 = pd.read_csv(StringIO(data2), escapechar='\\')
print(df_12)
```

```python
set4 = {1,2,3,4}
set5 = {2,3,4,5,6}

print(set4.difference(set5))
print(set4)

# print(set4.difference_update(set5))  #This returns None, as difference update doesnt give the final answer, while difference does..
set4.difference_update(set5)
print(set4)
```

```python
#Dictionaries
```

```python
#Creating an empty dictionary(2 ways)

dict1 = {}
print(type(dict1))

dict2 = dict()
print(type(dict2))
```

```python
#Revenue of companies(dictionary example)

dict3 = {"TCS" : 67, "Infosys" : 670, "Google" : 67000}
print(dict3)
```

```python
#Indexing + Appending + Updating_value

#Indexing(indexes are the keys instead of 0,1,2...)
print(dict3['TCS'])

#Appending
dict3['Apple'] = 6700
print(dict3)

#Updating
dict3['TCS'] = 670
print(dict3)
```

```python
#Traversing through the dictionary

#1-- keys

#M1
for key in dict3:
    print(key)

print('\n')

#M2
for key in dict3.keys():
    print(key)

print('\n')

#2-- values
for value in dict3.values():
    print(value)

print('\n')

#3-- items
for item in dict3.items():
    print(item)
```

```python
#Nested Dictionaries

dict_val1 = {'Mercedes' : 1960}
dict_val2 = {'Audi' : 1970}
dict_val3 = {'Ferrari' : 1980}

dict_cars = {'Car1' : dict_val1, 'Car2' : dict_val2,'Car3' : dict_val3}
print(dict_cars)
```

```python
#Indexing in Nested Dictionaries

print(dict_cars['Car2'])
print(dict_cars['Car2']['Audi'])
# print(dict_cars['Car2']['Mercedes']) This gives a key error
```
